```python
df.isnull().mean()
```

```python
missing_values_mean = df.isnull().sum()
```

```python
drop_from_data = missing_values_mean[missing_values_mean > 0.5].index
df = df.drop(drop_from_data, axis = 1 )
```

```python
df.head()
```

```python
cols_to_fill_median = missing_values_mean[(missing_values_mean < 0.5) & (missing_values_mean > 0)].index
df[cols_to_fill_median] = df[cols_to_fill_median].fillna(df[cols_to_fill_median].median())
```
